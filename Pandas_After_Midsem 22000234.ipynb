```python
left = pd.DataFrame({'key' : ['K0' , 'K1' , 'K2' , 'K3'],
                   'B' : ['B0' , 'B1' , 'B2' , 'B3'],
                   'C': ['C0' , 'C1' , 'C2' , 'C3']},
                    index=[0,1,2,3])
left
```

```python
right = pd.DataFrame({'key' : ['K0' , 'K1' , 'K2' , 'K3'],
                       'D': ['D0' , 'D1' , 'D2' , 'D3'],
                     'E': ['E0' , 'E1' , 'E2' , 'E3']},
                    index=[4,5,6,7])
right
```

```python
pd.merge(left , right , how="inner" , on = "key")
```

```python
pd.merge(left , right , how="outer" , on = "key")
```

```python
left = pd.DataFrame({'B' : ['B0' , 'B1' , 'B2' , 'B3'],
                   'C': ['C0' , 'C1' , 'C2' , 'C3']},
                    index=[0,1,2,3])
left
```

```python
right = pd.DataFrame({'D': ['D0' , 'D1' , 'D2' , 'D3'],
                     'E': ['E0' , 'E1' , 'E2' , 'E3']},
                    index=[0,0,2,3])
right
```

```python
left.join(right , how = 'outer')
```

```python
left.join(right , how = 'inner')
```
