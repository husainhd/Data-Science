```python
left=pd.DataFrame({'Key':['K0','K1','K0','K3'],'A':['A0','A1','A2','A3'],'B':['B0','B1','B2','B3']})
```

```python
right=pd.DataFrame({'Key':['K0','K3','K2','K3'],'C':['C0','C1','C2','C3'],'D':['D0','D1','D2','D3']})
```

```python
display(left,right,pd.merge(left,right,how='left',on='Key'),pd.merge(left,right,how='right',on='Key'),pd.merge(left,right,how='inner',on='Key'),pd.merge(left,right,how='outer',on='Key'))
```

```python
left1=pd.DataFrame({'A':['A0','A1','A2','A3'],'B':['B0','B1','B2','B3']},index=['K0','K1','K2','K0'])
right1=pd.DataFrame({'C':['C0','C1','C2','C3'],'D':['D0','D1','D2','D3']},index=['K0','K1','K2','K1'])
```

```python
left1.join(right1,how='left')
```

```python
left1.join(right1,how='right')
```

```python
left1.join(right1,how='inner')
```

```python
left1.join(right1,how='outer')
```
